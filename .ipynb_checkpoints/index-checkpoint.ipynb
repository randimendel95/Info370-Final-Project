## Introduction to K-Nearest-Neighbors (KNN)

K-nearest-neighbors (KNN) is a robust algorithm used to classify data. KNN falls in the supervised learning family of algorithms. In order to use KNN, we are given a labelled dataset consisting of training observations (x, y) and would like to capture the relationship between x and y. In this example we will use x to denote a _feature_ (i.e., an attribute, what we are using to predict) and y to denote the _outcome_ we are trying to predict.

In the classification setting, the K-nearest neighbor algorithm essentially boils down to forming a majority vote between the K most similar instances to a given “unseen” observation. One of the pros of using a KNN model is that it can learn complex decision boundaries on the fly. However, it is worth noting that the minimal training phase of KNN comes both at a _memory cost_, since we must store a potentially huge data set, as well as a _computational cost_ during test time since classifying a given observation requires a run down of the whole data set. 

## Apply KNN to Our Dataset


Now that we have explained KNN is is important to explain how that links with our data. We ended up with 5,791 universities which were used to create our model. We split this data into 80% training, and 20% test data, then ran 10-fold cross validation using a k nearest neighbors algorithm to find the universities that had the most similar attributes to a particular university in question, and tested this with multiple values of the most similar universities, by averaging their crime rates. We used 5 features to measure similarity between universities: percentage of financially independent students, percent of first generation university students, average family income, control of the institution(private or public), and the predominant degree achieved by students. Once we normalized these features from the minimum value to the maximum on a scale of 0 to 1, we used our algorithm to find the sum of distances of each feature between each university in the training set to our test value, in order to find the values with the least difference in feature to the university in question. We ran this using several different numbers of neighbors, and found the root mean squared error in our predictions for each test  set in our 10 folds, and found the average error in our predictions, choosing the number of compared neighbors with the least error in prediction.

As shown below, somewhere around **250 neighbors** would be our ideal K neighbors, however this model may not be accurate since this large number means that it is likely that neighbors with similar attributes don't actually tend to have the same crime rates, based on our given data, with only the given attributes weighted equally.

## Limitations

Our dataset had some inherent limitations, the greatest of which is the fact that it relies on reports from different sources. In some cases crime incident reports may have come from schools self-reporting, while in others they may have come from local law enforcement. In some instances schools did not report data. Additionally, schools are located in different states and local jurisdictions which may have different definitions of criminal offences. For instance, substances which may result in a drug-related criminal arrest at universities located in one state may be legal in others. 

Our dataset was also relatively small (~5,000 rows) for the type of predicting we aimed to do. We may have been able to predict more accurately with additional data. Lastly, we only analyzed five different features to predict crime rate. We therefore acknowledge that the predictions we made may not be accurate as there are many other factors that come into play.
